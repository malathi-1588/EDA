#  match

```python
import re
s1 = "python is my fav language"
pattern = 'python'
x = re.match(pattern,s1)
print(x)
```

```python
pattern = 'Python'
x = re.match(pattern,s1)
print(x)
```

# re.I --> ignore case

```python
pattern = 'python'
x = re.match(pattern,s1,re.I)
print(x)
print(x)
```

#  Search

```python
s2 = 'i like python.i use it to build applications'
x = re.search(pattern,s2,re.I)
print(x)
print(x.group)
```

# Findall

```python
y = re.findall(pattern,s2)
print(y)
```

# split

```python
s = re.split("\s",s2,3)
```

```python
print(s)
```

```python
s = re.split("\s",s2)
print(s)
```

# substitute

```python
re.sub("\s","-",s2)
```

```python
re.subn("\s","-",s2) ###integer returned - no of replacements made
```

# compile 

```python
pattern = '\d\d\d\d\d-\d\d\d\d\d'
s3 = '99406-84159 , 11111-22222'
c = re.compile(pattern)
```

```python
y = c.search(s3)

print(y)
```

```python
y = c.findall(s3)
y
```

```python
s = '12Python123'
c = re.compile(r'[a-zA-Z0-9]')
s1 = c.match(s)
print(s1)
```

```python
s1 = c.search(s)
s1
```

```python
s1 = c.findall(s)
print(s1)
print(''.join(s1))
```

```python
s = '12Python123@!!'
c = re.compile(r'[^a-zA-Z0-9]')
s1 = c.findall(s)
print(s1)
```

```python
pattern = "^a(b+)$"
s = 'a'
print(re.search(pattern,s))
```

```python
pattern = "^a(b*)$"
s = 'a'
print(re.search(pattern,s))
```

```python
pattern2 = "^[A-Z*]+[a-z*]+$"
s = 'Ab'
print(re.search(pattern2,s))
```

# # request response format

```python
import requests

from bs4 import BeautifulSoup

data = requests.get(url)
print(data) ## response starts with 4 series - client error, 5 series - server error
print(data.content)
```

```python
soup = BeautifulSoup(data.content,'html.parser')
```

```python
news_data = [{'headline':headline.find('span',attrs = {'itemprop':'headline'}).string,
             'article':article.find('div',attrs = {'itemprop':'articleBody'}).string
             }
     for headline,article in zip(soup.find_all('div',class_ = ["news-card-title news-right-box"]),
    soup.find_all('div',class_=["news-card-content news-right-box"]))]


```
