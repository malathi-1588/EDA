# g. What is the average tmdb_popularity for movies containing ‘Comedy’ as a genre?

```python
count = 0
s = 0

for i in range (len(data)):
    if("comedy" in data['genres'].iloc[i]):
        count += 1
        s += data['tmdb_popularity'].iloc[i]
s/len(data)
```

# h. In how many movie descriptions, the word ‘novel’ appears? 

```python
data = data.reset_index(drop=True)
c = 0
for i in range (len(data)):
    if("novel" in data['description'][i]):
        c += 1
c
```

# i. How many unique production countries are present in the data?

```python
uni = []

for i in range(len(data)):
    k = data['production_countries'][i]
    k = k[1:-1]
    k = k.split(',')
    for j in k:
        j = j.lstrip(" ")
        j = j.rstrip(" ")
        if j not in uni:
            uni += [j]
uni 
```

```python
len(uni)-1
```

# j. Average IMDB score for action movie – 7.22

```python
import numpy as np
data['action'] = np.nan
for i in range (len(data)):
    if("action" in data['genres'].iloc[i]):
        data['action'].iloc[i] = data['imdb_score'].iloc[i]
data['action'].mean()
```
