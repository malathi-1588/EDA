# Use matches.csv to answer the below questions.

```python
import pandas as pd
data1 = pd.read_csv(r'D:\Malathi\EDA\matches.csv')
data1.head()
```

# 1. Total number of tosses won by CSK?

```python
df1 = pd.pivot_table(data1, columns = 'team1', values = 'toss_winner', aggfunc = (lambda x:len(x)))
df1
```

```python
df1['Chennai Super Kings']
```

# 2. How many eliminator matches were held in Abu Dhabi and Dubai?

```python
df2 = pd.pivot_table(data1, columns = 'city', values = 'eliminator', 
               aggfunc = (lambda x:len(x)))
print(df2['Abu Dhabi'])
print(df2['Dubai'])
```

# 3. Who got the most Player of the match award after the year 2015 (2015 excluded)?

```python
from datetime import date
data1['date'] = pd.to_datetime(data1['date'])
after_2015 = data1[data1['date'] > pd.Timestamp(date(2015,12,31))]
after_2015
```

```python
obj = after_2015.groupby('player_of_match')
obj.first()
```

```python
obj.size()[obj.size()==max(obj.size())]
```

# 4. In how many matches Rohit Sharma (RG Sharma in the data) was awarded Player of the match after the year 2015?

```python
obj.size()['RG Sharma']
```
