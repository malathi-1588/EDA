# OUTLIERS
1. boxplot can be used to identify outliers in a column. 
2. The middle line in the boxplot gives info about how skewed the data is.
3. The values(dots) that lie above the upper line or below the lower line are the outliers
4. Boxplot is similar to normal distribution

5. IQR = Q3- Q1
6. upper = Q3 + 1.5 * IQR
7. lower = Q1 - 1.5 * IQR
8. where Q1 is the lower line of the box
   Q2 is middle line(green) of the box
   Q3 is upper line of the box

```python
import matplotlib.pyplot
data.boxplot(column='Age')
```

```python
def IQR(data, col):
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3 - q1
    return iqr, q1, q3
```

```python
iqr, q1, q3 = IQR(data, 'Age')
upper = q3+1.5*iqr
lower = q1-1.5*iqr
print("upper:",upper)
print('lower:',lower)
```

```python
data[(data['Age']>lower) & (data['Age']>=upper)]
```

# Merging
* to merge 2 different dataframes
* atleast one column shud be common to perform merging
* (inner, left, right, full-outer) types of merging

* df1                     df2
* A B C D                 A E F

1. inner gives intersection of dataframes
    all columns are present but only common rows in A are included
    
2. left gives all columns with unique and common rows but for unique values only data from left dataframe is included
    intersection + left

3. right gives all columns with unique and common rows but for unique values only data from rigth dataframe is included
    intersection + right
    
4. full outer
    intersection + right + left


```python
import pandas as pd
df1 = pd.read_csv(r"D:\Malathi\EDA\studentData.csv")
df2 = pd.read_excel(r"D:\Malathi\EDA\current_Students.xlsx")
```

```python
print(len(df1), len(df2))
```

```python
#on is a parameter, based on which mergeing is done
#how is a parameter that take either inner, outer or left syntax: how='inner'
df = pd.merge(df2, df1, left_on = 'application_no', right_on = 'AppNo' )
df
```

# Binning
* categorizing, discretize
* Ex age can be categorised into 18-25 youth, 25-40 mid age, 40-60 adult,
* 18-25, 25-40, 40-60 are limits of each bin
* We shud know the min and max values of the column before customising the bins
* (18, 30], here 30 is inclusive, in the next category it starts from 31
* upper limit is inclusive
* if the value does not fit in any of the customised bins then NaN is showed in the output, this problem does not occur in the bins created by python itself
* 

```python
import pandas as pd
ads = pd.read_csv(r'D:\Malathi\EDA\ads.csv')
```

# Encoding
1. Label Encoding
2. One-hot Encoding
* ML models cannot process string data, hence the ads data with the age_category cannot be processed

* label encoding is used when the data is of ordinal type (ordinal data - some hierarchy is present in the data)
  Ex: youth < mid age < adult
* Mapping the category to a numerical value
  1. age           label
  2. youth         0
  3. mid age       1
  4. adult         2

* one-hot encoding is used for nominal data (nominal data - no hierarchy is present in the data)

```python
ads_noage = ads.drop('Age', axis = 1)
ads_noage
```
